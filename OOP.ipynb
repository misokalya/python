```python
#Simple Calculator

class Calculator:
    def add(self, a, b):
        return a + b
    
    def multiply(self, a, b):
        return a*b
    
    def subtract(self, a, b):
        return a-b
    
    def divide(self, a, b):
        if b == 0:
            return "Can not divide by zero!"
        return a/b
    

a = int(input("Enter first number: "))
b = int(input("Enter the second number: "))

calc = Calculator()
print("Add: ", calc.add(a,b))
print("Multiply: ", calc.multiply(a,b))
print("Subtract: ", calc.subtract(a,b))
print("Divide: ", calc.divide(a,b))
```

```python
# __str__() method

class Person:
    def __init__ (self, name, age):
        self.name = name
        self.age = age

    def __str__ (self):
        return f"{self.name} is {self.age}"
    
p1 = Person("Miso", 100)
print(p1)
```

```python
#Multitple Methods

class Playlist:
    def __init__ (self, name):
        self.name = name
        self.songs = []

    def add_song(self, song):
        self.songs.append(song)
        print(f"Added: {song}")
    
    def remove_song(self, song):
        if song in self.songs:
            self.songs.remove(song)
            print(f"Removed: {song}")
    
    def show_songs(self):
        print(f"Playlist '{self.name}'")
        for song in self.songs:
            print(f"{song}")

my_playlist = Playlist("Miso's Playlist")

my_playlist.add_song("Song 1 - Song name")
my_playlist.add_song("Song 2 - Song name")
print("============================")
my_playlist.show_songs()
```
