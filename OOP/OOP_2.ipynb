```python
#Inheritance

class Person():
    def __init__ (self, firstname, lastname):
        self.firstname = firstname
        self.lastname = lastname
    
    def printname(self):
        print(self.firstname, self.lastname)
    
p1 = Person("Jovline", "Dela Cruz")
p1.printname()

class Student(Person):
    pass

x = Student("Mike", "Olsen")
x.printname()
```

```python
#Polymorphism

class Car:
    def __init__(self, brand, model):
        self.brand = brand
        self.model = model
    
    def move(self):
        print("Drive!")
    
class Boat:
    def __init__(self, brand, model):
        self.brand = brand
        self.model = model
    
    def move(self):
        print("Sail!")

class Plane:
    def __init__(self, brand, model):
        self.brand = brand
        self.model = model

    def move(self):
        print("Fly!")


car1 = Car("Toyota", "Corolla")
boat1 = Boat("Yamaha", "242X")
plane1 = Plane("Boeing", "747")

for x in (car1, boat1, plane1):
    x.move()
```

```python
#Encapsulation

class Person():
    def __init__ (self, name, age):
        self.name = name
        self.__age = age #private attribute
    
    def get_age(self):
        return self.__age
    
p1 = Person("Jovline", 25)
print(p1.name)
print(p1.get_age())
```
